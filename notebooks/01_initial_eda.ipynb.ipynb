```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Define the path to the data
data_path = '../data/creditcard.csv'

# Load the dataset
df = pd.read_csv(data_path)
```

```python
# Display the shape of the dataframe
print(f"Dataset shape: {df.shape}")

# Display the first 5 rows
print("\nFirst 5 rows:")
display(df.head())

# Display data types and non-null counts
print("\nData Info:")
df.info()
```

```python
# Display summary statistics for numerical columns
print("\nSummary Statistics:")
display(df.describe())
```

```python
# Check for null values in each column
print("\nNull Value Counts:")
print(df.isnull().sum())
```

```python
# Check the distribution of the 'Class' target variable
class_counts = df['Class'].value_counts()
print("Class Distribution:")
print(class_counts)

# Visualize the imbalance
plt.figure(figsize=(8, 6))
sns.countplot(x='Class', data=df)
plt.title('Distribution of Fraudulent vs. Non-Fraudulent Transactions')
plt.xlabel('Class (0: Non-Fraud, 1: Fraud)')
plt.ylabel('Count')
plt.xticks([0, 1], ['Non-Fraud', 'Fraud'])
plt.show()

# Print the percentage of fraudulent transactions
fraud_percentage = (class_counts[1] / class_counts.sum()) * 100
print(f"\nPercentage of fraudulent transactions: {fraud_percentage:.4f}%")
```
